# Q1: Data Analyst position in Banglore (www.naukri.com)

```python
title=[]
location=[]
company=[]
experience=[]
```

```python
for i in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]"):
    location1=i.text
    location.append(location1)
location[0:10]
```

```python
for i in driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']"):
    companys=i.text
    company.append(companys)
company[0:10]
    
```

```python
for i in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]"):
    experiences=i.text
    experience.append(experiences)
experience[0:10]
```

```python
dajobs=pd.DataFrame({})
dajobs['job-title']=title[0:10]
dajobs['job-location']=location[0:10]
dajobs['company_name']=company[0:10]
dajobs['experience_required']=experience[0:10]
dajobs
```

# Q6 100 Sunglasses from Flipkart

```python
import selenium
from selenium import webdriver
import pandas as pd
driver=webdriver.Chrome("chromedriver.exe")
driver.get('https://www.flipkart.com/')
```

```python
search=driver.find_element_by_xpath("//input[@class='_3704LK']")
search.send_keys('Sunglasses')
button=driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
```

```python
Brand=[]
Description=[]
Price=[]
Discount=[]
start=0
end=5

for page in range(start,end+1):
    Name=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    for i in Name:
        brand=i.text
        Brand.append(brand)
    for i in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
        description=i.text
        Description.append(description)
    for i in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
        price=i.text
        Price.append(price)
    for i in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span[1]"):
        discount=i.text.replace('off','')
        Discount.append(discount)
    
        nxt=driver.find_element_by_xpath("//a[@class='_1LKTO3']/span[1]")
        if nxt.text=='Next':
            nxt.click()
        
Sunglasses=pd.DataFrame({'Brand_Name':Brand[0:100],'Product_Description':Description[0:100],'Price':Price[0:100],'Discount':Discount[0:100]})
Sunglasses
    
```
